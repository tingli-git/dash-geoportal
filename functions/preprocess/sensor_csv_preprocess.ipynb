```python
import pandas as pd
import numpy as np
import glob
```

```python
top_dir = "/datawaha/esom/DatePalmCounting/Geoportal/Sensors/SensorReads"
files = glob.glob(f"{top_dir}/*.csv")
```

```python
for file in files:
    df = pd.read_csv(file)
    print(f"Processing file: {file} with shape {df.shape}")
    # Add your preprocessing steps here
    # get the soil moisture for zoot zone
    # if A(all) in df.columns drop it 
    if 'A(all)' in df.columns:
        df = df.drop(columns=['A(all)'])
    # find all columns that contains "A"
    a_columns = [col for col in df.columns if 'A' in col]
    # calculate the sum of these columns row-wise and divide by 9
    df['soil_moisture_root_zone'] = df[a_columns].sum(axis=1) / 9
    # now get the daily average soil moisture
    # save to the same dataframe
    df.to_csv(file.replace('.csv','_v2.csv'), index=False)
    # 
```
