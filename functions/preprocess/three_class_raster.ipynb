```python
import os 
import rioxarray as rxr
import xarray as xr
```

```python
top_dir = '/datawaha/esom/DatePalmCounting/AncillaryData/Deeplearning/predictions/RF/predi_raster' 
vege_novege_path = f"{top_dir}/yhat_labels.tif"
tree_notree_path = f"{top_dir}/tree_nontree_yhat_labels.tif"
# now read these two files
vege_novege = rxr.open_rasterio(vege_novege_path, masked=True).squeeze()
tree_nontree = rxr.open_rasterio(tree_notree_path, masked=True).squeeze()
```

```python
# now update the tree_nontree by updating the pixels to -1 if vege_novege is 0 
tree_nontree = tree_nontree.where(vege_novege==1, other=-1) 
# now add 2 to tree_nontree to make the classes 1,2,3
tree_nontree = tree_nontree + 2
# now save the final raster
output_path = f"{top_dir}/three_class_raster.tif"
tree_nontree.rio.to_raster(output_path, compress='LZW', dtype='int8')
```
