```python
import geopandas as gpd
import numpy as np, pandas as pd
```

```python
gpd_all = gpd.read_file('/datawaha/esom/DatePalmCounting/Geoportal/Datepalm/app_server/datepalms/Qassim_datepalm_fields_polygons.shp')
```

```python
gpd_all['Field_id'] = np.arange(1, len(gpd_all)+1)
```

```python
gpd_all.columns
```

```python
# remove unnecessary columns
gpd_all= gpd_all[[ 'Field_id', 'geometry','year_check']]
try:
    gpd_all = gpd_all.set_crs(epsg=4326)
except Exception:
    pass
# reproject to UTM Zone 38N
gpd_all = gpd_all.to_crs(epsg=32638)
# get the field area in square meters
gpd_all['Field_Acreage_m2'] = gpd_all['geometry'].area
gpd_all['Field_Acreage_ha'] = gpd_all['Field_Acreage_m2'] / 10000.0
# round to 2 decimal places
gpd_all['Field_Acreage_ha'] = gpd_all['Field_Acreage_ha'].round(2)

gpd_all = gpd_all.dropna(subset=['Field_Acreage_m2'])

gpd_all['Rough_Tree_Count'] = (
    gpd_all['Field_Acreage_m2'] / 100
).astype(int)

# now back to WGS84
gpd_all = gpd_all.to_crs(epsg=4326)
# export to geojson
gpd_all.to_file('/datawaha/esom/DatePalmCounting/Geoportal/Datepalm/app_server/datepalms/Qassim_datepalm_fields_polygons.geojson', driver='GeoJSON')
```
