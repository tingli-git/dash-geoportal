```python
from ipyleaflet import Map, TileLayer, LayersControl, ScaleControl
from ipywidgets import Layout

m = Map(center=(26.4, 43.8), zoom=8, layout=Layout(height="10vh", width="100%"))

# ESRI base (optional)
m.add_layer(TileLayer(
    url="https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}",
    name="ESRI World Imagery"
))

# Your XYZ via local HTTP
m.add_layer(TileLayer(
    url="http://localhost:8766/{z}/{x}/{y}.png",
    name="Test XYZ Overlay",
    opacity=1.0,
    no_wrap=True
))
```

```python
m.add_control(LayersControl(position="topright"))
m.add_control(ScaleControl(position="bottomleft"))
m
```
